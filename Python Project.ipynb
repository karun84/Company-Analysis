1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
distribution=data['Team'].value_counts()
percentSplit=(distribution/len(data))*100
percentSplit
```

2. Segregate employees based on their positions within the company.

```python
data["Salary"]=data["Salary"].replace('NaN','')
```

```python
data["Salary"].fillna(data["Salary"].mean())
```

```python
data['College'].fillna('Unknown')
```

3. Identify the predominant age group among employees.

```python
bins = [20, 25, 30, 35, 40, 45]
labels = ['20-25', '26-30', '31-35', '36-40', '41-45']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
ageGroup=data['Age Group'].value_counts()
predominant_age_group = ageGroup.idxmax()
predominant_count = ageGroup.max()
print(f"The predominant age group is '{predominant_age_group}' with {predominant_count} employees.")
```

4. Discover which team and position have the highest salary expenditure

```python
salary_expenditure = data.groupby(['Team', 'Position'])['Salary']
highest_expenditure=salary_expenditure.idxmax()
highest_expenditure
```

5. Investigate if there's any correlation between age and salary, and represent it visually

```python
correlation=data['Age'].corr(data['Salary'])
print(f"Correlation between age and salary is:{correlation}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Salary'])
#sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees

```python
distribution=data['Team'].value_counts()
percentSplit=(distribution/len(data))*100
plt.figure(figsize=(8, 8))
plt.pie(percentSplit, labels=distribution.index,autopct='%1.1f%%', startangle=140)
plt.title('Team Splited by percentage')
plt.show()
```

Discover which team and position have the highest salary expenditure. 

```python
salary_expenditure= data.groupby('Team')['Salary'].sum().sort_values()
plt.figure(figsize=(12,6))
plt.barh(salary_expenditure.index, salary_expenditure.values, color='salmon', edgecolor='black')
plt.title('Total salary by team')
plt.xlabel('Total salary')
plt.ylabel('Team')
plt.show()
```
